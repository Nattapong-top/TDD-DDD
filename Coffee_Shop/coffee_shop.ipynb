# ☕ บทที่ 1: TDD ร้านกาแฟ (สเต็ป Red Phase สร้างเทสต์ให้พัง)
---

ในสเต็ปนี้เราจะทำตามคลิปเป๊ะๆ โดยแยกไฟล์ระหว่าง **โค้ดทดสอบ (Test)** และ **โค้ดใช้งานจริง (Logic)** ออกจากกัน การทำ TDD เราต้องเริ่มจากเขียนเทสต์ก่อนเสมอครับ

| คำศัพท์ (Technical Term) | คำอ่านไทย (Pronunciation) | ความหมาย (Translation/Definition) |
| :--- | :--- | :--- |
| **Test Case** | เทสต์-เคส | **กรณีทดสอบ** (เงื่อนไขที่เราสร้างขึ้นมาเพื่อพิสูจน์ว่าระบบทำงานถูกต้อง) |

| **Red Phase** | เรด-เฟส | **สถานะพัง** (จุดเริ่มต้นของ TDD คือการรันเทสต์ให้พัง เพื่อให้แน่ใจว่าเทสต์ทำงานได้จริง) |

# 🧱 บทที่ 1.3: สร้างรากฐานด้วย Value Object (ควรเขียนตอนไหน?)
---

คำถามนี้คมมากครับป๋า! ในคลิปวิดีโอที่เขาสอน เขาไม่ได้เขียน Value Object (VO) เพราะโจทย์เน้นความเข้าใจ TDD แบบพื้นฐานที่สุด จึงใช้ Primitive Type (`str`, `float`) ไปก่อน แต่สำหรับโลกของ **DDD (Domain-Driven Design)** และ **Clean Architecture** ที่ป๋ากำลังฝึกปรืออยู่นั้น...

คำตอบคือ: **"เราควรเขียน Value Object เป็นอันดับแรกสุด (Step 0) ก่อนที่จะไปเขียน Entity ครับป๋า!"**

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **Value Object (VO)** | แวลู อ็อบเจกต์ | **อ็อบเจกต์ห่อหุ้มข้อมูล** (ไม่มี ID ของตัวเอง สนใจแค่ค่าที่อยู่ข้างใน และต้องตรวจสอบความถูกต้องของตัวเองได้) |
| **Primitive Obsession** | พริม-มิ-ทีฟ ออบ-เซส-ชัน | **การเสพติดตัวแปรพื้นฐาน** (Anti-pattern ที่ใช้ `int`, `str`, `float` มาแทนคอนเซปต์ทางธุรกิจ ทำให้โค้ดพังง่าย) |
| **Encapsulation** | เอ็น-แคป-ซู-เล-ชัน | **การแคปซูลข้อมูล** (การห่อหุ้มข้อมูลและเงื่อนไขการตรวจสอบไว้ด้วยกันในที่เดียว) |

---

### 🤔 ทำไมต้องสร้าง Value Object ก่อน Entity?

ถ้าป๋าไม่สร้าง VO ก่อน ภาระการตรวจสอบข้อมูลทั้งหมด (เช่น เช็คว่าเงินติดลบไหม, เช็คว่าชื่อเครื่องดื่มว่างเปล่าไหม) จะไปตกหนักอยู่ที่ Entity (`Cafe`) ทำให้ Entity บวมและโค้ดรกครับ

ในทางกลับกัน ถ้าเราสร้าง VO ขึ้นมาทำหน้าที่ "เฝ้าประตู" ไว้ตั้งแต่แรก Entity ของเราจะสนใจแค่ **Business Logic หลัก** (เช่น การหักลบเงินทอน) โดยไม่ต้องกังวลว่าข้อมูลที่รับเข้ามาจะผิดพลาดเลยครับ

### 🛠️ TDD Step-by-Step: การสร้าง Value Object `Money` ด้วย Pydantic

เราจะมาประยุกต์ใช้หลักการ TDD (Red-Green-Refactor) เพื่อสร้าง `Money` กันก่อนเลยครับ

```python

# ไฟล์: tests/test_value_objects.py
import pytest
from pydantic import ValidationError
# from domain.value_objects import Money  <-- เดี๋ยวเราจะไปสร้างตัวนี้กัน

# ---------------------------------------------------------
# 1. เขียน Test ให้ Value Object (Red Phase)
# ---------------------------------------------------------

def test_money_cannot_be_negative():
    # [Arrange & Act & Assert]
    # คาดหวังว่าถ้าเผลอสร้างเงิน "ติดลบ" ระบบจะต้องเตะ Error (ValidationError) ออกมาทันที
    with pytest.raises(ValidationError):
        invalid_money = Money(amount=-10.0)

def test_money_can_be_created_with_valid_amount():
    # [Arrange & Act] สร้างเงิน 50 บาท
    valid_money = Money(amount=50.0)

    # [Assert] ค่า amount ต้องเท่ากับ 50.0 เป๊ะๆ
    assert valid_money.amount == 50.0

````python
# ไฟล์: domain/value_objects.py
from pydantic import BaseModel, Field

# ---------------------------------------------------------
# 2. สร้าง Value Object ให้ Test ผ่าน (Green Phase)
# ---------------------------------------------------------

class Money(BaseModel):
    """
    Value Object สำหรับจัดการเรื่องเงิน
    Business Logic: จำนวนเงินในระบบห้ามมีค่าติดลบเด็ดขาด (ge=0.0 คือ Greater than or Equal to 0)
    """
    amount: float = Field(..., ge=0.0, description="จำนวนเงินต้องไม่ติดลบ")

class DrinkName(BaseModel):
    """
    Value Object สำหรับชื่อเครื่องดื่ม
    Business Logic: ห้ามเป็น String ว่างๆ ต้องมีตัวอักษรอย่างน้อย 1 ตัว
    """
    value: str = Field(..., min_length=1, description="ชื่อเครื่องดื่มห้ามว่าง")
````

````python
# ไฟล์: domain/logic.py
# from domain.value_objects import DrinkName, Money

class Cafe:
    def __init__(self):
        # ราคาสินค้าก็ควรเก็บเป็น Value Object เช่นกัน
        self.PRICE = {
            "Tea": Money(amount=20.0),
            "Coffee": Money(amount=30.0)
        }

    # ห้ามใช้ drink: str, money: float ตรงๆ! ต้องบังคับรับเป็น Value Object
    def buy(self, drink: DrinkName, payment: Money) -> tuple[DrinkName, Money]:

        drink_price = self.PRICE.get(drink.value)
        if not drink_price:
            raise ValueError("Drink not in menu")

        # 1. Guard Clause: เช็คเงินพอไหม? (ข้อมูลที่เข้ามาการันตีแล้วว่าไม่ติดลบ)
        if payment.amount < drink_price.amount:
            raise ValueError("Money not enough")

        # 2. Calculate: คำนวณเงินทอน
        change_amount = payment.amount - drink_price.amount

        # 3. ห่อกลับเป็น Value Object ก่อนส่งคืนออกไป
        change = Money(amount=change_amount)

        return drink, change
````

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **Syntax Error** | ซิน-แท็กซ์ เออร์-เรอร์ | **ไวยากรณ์ผิดพลาด** (การพิมพ์โค้ดผิดกฎของภาษา เช่น ลืมพิมพ์ `def`, ลืมเครื่องหมาย `:`) |
| **Typo** | ไท-โป | **พิมพ์ผิด/พิมพ์ตกหล่น** (เรื่องธรรมชาติที่โปรแกรมเมอร์ทุกคนต้องเจอทุกวัน) |
| **Absolute Import** | แอ๊บ-โซ-ลูท อิม-พอร์ต | **การนำเข้าไฟล์แบบระบุตำแหน่งเต็ม** (เช่น `from domain.value_object import ...`) |

### 🚀 สเต็ปแก้ปัญหา: ก๊อปปี้ทับไปเลยครับป๋า!

เพื่อให้ป๋าไปต่อได้แบบลื่นๆ จุกทองเตรียมโค้ดที่ถูกต้องสมบูรณ์แบบของทั้ง 3 ไฟล์มาให้แล้วครับ ป๋าแค่ก๊อปปี้ไปวางทับในไฟล์ของป๋าได้เลยครับ โดยยึดโครงสร้างโฟลเดอร์ตามนี้ครับ:

`<โครงสร้างโฟลเดอร์>`
```text
Coffee_Shop/
├── domain/
│   ├── __init__.py
│   ├── shop_logic.py     # ไฟล์ที่ 2
│   └── value_object.py   # ไฟล์ที่ 1
└── tests/
    ├── __init__.py

📁 ไฟล์ที่ 1: domain/value_object.py
(ไฟล์นี้เก็บ Value Object ที่คอยปกป้องข้อมูลให้เรา)

````python
from pydantic import BaseModel, Field

class DrinkName(BaseModel):
    """Value Object สำหรับชื่อเครื่องดื่ม"""
    value: str = Field(..., min_length=1)

class MoneyTHB(BaseModel):
    """Value Object สำหรับเงินบาท (ห้ามติดลบ)"""
    amount: float = Field(..., ge=0.0)
````


📁 ไฟล์ที่ 2: domain/shop_logic.py
(ก๊อปปี้ไปทับไฟล์เดิมที่พังอยู่ได้เลยครับ แก้คำผิดและ Import ให้ครบแล้ว)

````python

# นำเข้า Value Object ที่สร้างไว้มาใช้งาน
from domain.value_object import DrinkName, MoneyTHB

class Shop:
    def __init__(self):
        # กำหนดราคาสินค้าโดยห่อด้วย Value Object
        self.PRICE = {
            'Tea': MoneyTHB(amount=20.0),
            'Coffee': MoneyTHB(amount=30.0)
        }

    def buy(self, drink: DrinkName, payment: MoneyTHB) -> tuple[DrinkName, MoneyTHB]:
        # ดึงราคาจาก Dictionary
        drink_price = self.PRICE.get(drink.value)

        # Guard Clause 1: เช็คว่ามีเมนูนี้ไหม
        if drink_price is None:
            raise ValueError("Drink not in menu")

        # Guard Clause 2: เช็คว่าเงินพอไหม
        if payment.amount < drink_price.amount:
            raise ValueError("Money not enough")

        # คำนวณเงินทอน (Business Logic หลัก)
        change_amount = payment.amount - drink_price.amount

        # ห่อผลลัพธ์กลับเป็น Value Object ก่อนคืนค่า
        return drink, MoneyTHB(amount=change_amount)
````


📁 ไฟล์ที่ 3: tests/test_coffee_shop.py
(ไฟล์นี้สำหรับเขียน Test Case เพื่อรันพิสูจน์ระบบครับ)

````python
from domain.shop_logic import Shop
from domain.value_object import DrinkName, MoneyTHB

def test_should_return_tea_and_change_30_when_buy_tea_with_50():
    # [Arrange] เตรียมระบบ
    shop = Shop()
    order = DrinkName(value="Tea")
    pay = MoneyTHB(amount=50.0)

    # [Act] กระทำ Business Logic: สั่งซื้อ
    drink, change = shop.buy(order, pay)

    # [Assert] ตรวจสอบผลลัพธ์
    assert drink.value == "Tea"
    assert change.amount == 30.0
````

# 🎯 บทที่ 1.4: สร้าง Custom Exception แบบต้นฉบับเป๊ะๆ
---

ตาไวมากครับป๋า! จุกทองเห็นจากรูปที่ 3 แล้ว ต้นฉบับเขาใช้เทคนิคการสร้าง **Custom Exception** (`DrinkNotInMenu`) และแยกการตรวจสอบไปไว้ในฟังก์ชันส่วนตัว (Private Method) ที่ชื่อ `_validate_drink` เพื่อให้โค้ดในฟังก์ชัน `buy` อ่านง่ายและเป็นระเบียบที่สุดครับ

การทำแบบนี้ดีมากๆ เพราะเวลาที่ระบบ Error มันจะไม่ฟ้องแค่ `ValueError` ทั่วไป แต่จะฟ้องชื่อ Error ที่สื่อความหมายทางธุรกิจชัดเจนเลยครับ เรามาปรับโค้ดของเราให้เหมือนต้นฉบับ โดยยังคงความแข็งแกร่งของ Value Object ไว้กันครับ!

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **Custom Exception** | คัส-ตอม เอ็ก-เซป-ชัน | **ข้อผิดพลาดที่สร้างขึ้นเอง** (การสร้าง Class Error เฉพาะเจาะจงสำหรับ Business ของเรา) |
| **Private Method** | ไพร-เวต เมท-ธอด | **ฟังก์ชันใช้งานภายใน** (ฟังก์ชันที่ขึ้นต้นด้วย `_` เพื่อบอกว่าฟังก์ชันนี้ใช้เรียกกันเองภายใน Class เท่านั้น ห้ามคนนอกเรียกใช้) |
| **pytest.raises** | พาย-เทสต์ เรส-เซส | **การดักจับ Error ใน Test** (คำสั่งสำหรับทดสอบว่าโค้ดของเราทำการพ่น Error ออกมาตามที่คาดหวังหรือไม่) |

> **💡 ข้อคิดจากจุกทอง:**
> การตั้งชื่อ Exception ให้สื่อถึง Business Logic (เช่น `DrinkNotInMenu` หรือ `MoneyNotEnough`) คือหัวใจสำคัญของ DDD ครับ มันทำให้โค้ดของเรากลายเป็นเอกสาร (Self-documenting code) อ่านปุ๊บรู้ปั๊บว่าพังเพราะกฎข้อไหน!

---

### 🛠️ สเต็ป 1: สร้าง Custom Exception และปรับ Logic ให้เหมือนต้นฉบับ

ป๋าก๊อปปี้โค้ดด้านล่างนี้ ไปทับในไฟล์ `domain/shop_logic.py` ได้เลยครับ จุกทองเพิ่ม Class `DrinkNotInMenu` และฟังก์ชัน `_validate_drink` ให้เหมือนในคลิปเป๊ะๆ แล้วครับ


````python

# ไฟล์: domain/shop_logic.py
from domain.value_object import DrinkName, MoneyTHB

# ---------------------------------------------------------
# สร้าง Custom Exception (สืบทอดจาก Exception พื้นฐานของ Python)
# ---------------------------------------------------------
class DrinkNotInMenu(Exception):
    """Error: เมื่อลูกค้าสั่งเครื่องดื่มที่ไม่มีในเมนู"""
    pass

class MoneyNotEnough(Exception):
    """Error: เมื่อลูกค้าจ่ายเงินไม่พอ"""
    pass

# ---------------------------------------------------------
# Entity: Shop
# ---------------------------------------------------------
class Shop:
    def __init__(self):
        self.PRICE = {
            'Tea': MoneyTHB(amount=20.0),
            'Coffee': MoneyTHB(amount=30.0)
        }

    def _validate_drink(self, drink: DrinkName):
        # ตรวจสอบว่าชื่อเครื่องดื่ม (drink.value) มีอยู่ใน Dictionary หรือไม่
        if drink.value not in self.PRICE:
            # พ่น Custom Exception ตามต้นฉบับ
            raise DrinkNotInMenu()

    def buy(self, drink: DrinkName, payment: MoneyTHB) -> tuple[DrinkName, MoneyTHB]:
        # 1. เรียกใช้งาน Private Method เพื่อเช็คเมนู (ถ้าไม่มี จะเด้ง Error ออกไปเลย)
        self._validate_drink(drink)

        # 2. ดึงราคาออกมาคำนวณ (ถึงบรรทัดนี้มั่นใจได้ว่ามีเมนูแน่นอน)
        drink_price = self.PRICE[drink.value]

        # 3. Guard Clause: เช็คเงินพอไหม?
        if payment.amount < drink_price.amount:
            raise MoneyNotEnough()

        # 4. คำนวณเงินทอน และห่อกลับเป็น Value Object
        change_amount = payment.amount - drink_price.amount
        return drink, MoneyTHB(amount=change_amount)
````

# 💉 บทที่ 1.4: Dependency Injection (DI) ฉีดของนอกเข้าข้างใน
---

มาถึงจุดที่เรียกได้ว่าเป็น "หัวใจสำคัญ" ของการทำ Clean Architecture และ TDD เลยครับป๋า! ในคลิปวิดีโอตอนที่เขาสอนทำเมนู **"Surprise" (สุ่มเครื่องดื่ม)** ป๋าจะเห็นว่าเขาเจอปัญหาคือ "ถ้าใช้คำสั่งสุ่ม (Random) ตรงๆ ในโค้ด เราจะเขียน Test ไม่ได้ เพราะเดาผลลัพธ์ไม่ได้!"

วิธีแก้ปัญหาระดับปรมาจารย์ที่เขาใช้ก็คือ **Dependency Injection (DI)** ครับ เรามาดูคำศัพท์และเจาะลึกวิชานี้กันครับ!

### 📚 ตารางคำศัพท์

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **Dependency Injection (DI)** | ดี-เพน-เดน-ซี อิน-เจก-ชัน | **การฉีดความพึ่งพา** (การส่งอ็อบเจกต์หรือฟังก์ชันที่ต้องใช้งาน เข้ามาจากภายนอก แทนที่คลาสจะสร้างขึ้นมาเอง) |
| **Dependency** | ดี-เพน-เดน-ซี | **สิ่งที่ต้องพึ่งพา** (เช่น ระบบร้านกาแฟ ต้องพึ่งพาระบบสุ่มเมนู ระบบสุ่มนั่นแหละคือ Dependency) |
| **Mocking / Mock Object** | ม็อก-กิง / ม็อก-อ็อบเจกต์ | **ของปลอมจำลอง** (อ็อบเจกต์หน้าม้าที่เราสร้างขึ้นมาเพื่อควบคุมผลลัพธ์ใน Test Case) |
| **Tight Coupling** | ไทต์ คัป-ปลิง | **การผูกมัดแน่นเกินไป** (การเรียกใช้คลาสหรือไลบรารีอื่นตรงๆ ทำให้แก้โค้ดยากและเทสต์ไม่ได้) |

---

### 🧠 หลักการทำงานของ DI (ทำไมถึงต้องฉีด?)



ลองนึกภาพตามจุกทองนะครับ... ถ้าป๋าสร้างคลาสร้านกาแฟ แล้วให้ร้านกาแฟ **"สร้างเครื่องสุ่มเมนูเอง"** ไว้ข้างใน (Tight Coupling) พอถึงเวลาที่เราจะทดสอบ (Test) เราจะไม่มีทางควบคุมเครื่องสุ่มนั้นได้เลย เพราะมันฝังตัวอยู่ข้างในร้านกาแฟ

**การทำ DI คือการเปลี่ยนแนวคิดใหม่:** เราจะออกแบบให้ร้านกาแฟ **"มีช่องเสียบ"** (ผ่านทางพารามิเตอร์หรือ Constructor) เพื่อให้เราสามารถ **"ฉีด (Inject)"** เครื่องสุ่มจากภายนอกเข้าไปได้
* ตอนใช้งานจริง (Production) -> เราฉีด "เครื่องสุ่มของจริง" เข้าไป
* ตอนเขียนเทสต์ (TDD) -> เราฉีด "เครื่องสุ่มหน้าม้า (Mock)" ที่ล็อคผลลัพธ์ได้ตามใจสั่งเข้าไปแทน!

---

### 💻 ตัวอย่างโค้ด: การกอบกู้สถานการณ์ด้วย DI

เรามาดูการจำลองเมนู Surprise จากวิดีโอกันครับ จุกทองจัดรูปแบบให้ดูสะอาดและมี Comment อธิบาย Business Logic ชัดเจนครับ

```python
# ไฟล์: domain/logic.py
from typing import Callable
from pydantic import BaseModel

class DrinkName(BaseModel):
    value: str

class Cafe:
    """
    Entity: ร้านกาแฟ
    เราออกแบบให้รับการฉีด (Inject) ฟังก์ชันสุ่มจากภายนอกผ่าน Constructor
    """
    def __init__(self, random_selector: Callable[[list[str]], str]):
        # เก็บฟังก์ชันที่ถูกฉีดเข้ามาไว้เป็น attribute
        self._random_selector = random_selector
        self.menu_list = ["Tea", "Coffee"]

    def buy_surprise(self) -> DrinkName:
        # Business Logic: เลือกระบบสุ่มเมนูจาก Dependency ที่ฉีดเข้ามา
        # แทนที่จะใช้คำสั่ง random.choice() ของ Python โดยตรง
        selected_drink_str = self._random_selector(self.menu_list)

        # ห่อหุ้มด้วย Value Object ก่อนส่งกลับ
        return DrinkName(value=selected_drink_str)

# 🏗️ บทที่ 1.5: ชำแหละ def __init__ (ประตูบานแรกของ Object)
---

ในโลกของการเขียนโปรแกรมเชิงวัตถุ (OOP) `__init__` คือ "จุดเริ่มต้นของทุกสิ่ง" ครับป๋า แต่พอเราเอามาใช้กับ **DDD (Domain-Driven Design)** และ **Clean Architecture** หน้าที่ของมันจะยิ่งทวีความสำคัญขึ้นไปอีกระดับเลยครับ!

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **__init__ (Constructor)** | อิน-นิต (คอน-สตรัก-เตอร์) | **ฟังก์ชันตั้งต้น** (ฟังก์ชันที่จะถูกเรียกทำงานอัตโนมัติทันทีที่เราสร้างอ็อบเจกต์ใหม่) |
| **self** | เซลฟ์ | **ตัวมันเอง** (คำจำกัดความที่อ็อบเจกต์ใช้เรียกตัวเอง เพื่อเข้าถึงตัวแปรหรือฟังก์ชันของมัน) |
| **Instantiation** | อิน-สแตน-ชิ-เอ-ชัน | **การถอดแบบ** (กระบวนการสร้างอ็อบเจกต์ (Object) ออกมาจากแม่พิมพ์ (Class)) |
| **Dependency Injection** | ดี-เพน-เดน-ซี อิน-เจก-ชัน | **การฉีดของเข้าทาง __init__** (การส่งมอบสิ่งที่คลาสต้องใช้ผ่านทางวงเล็บตอนสร้างคลาส) |



### 🛠️ การทำงานของ `__init__` ในมุมมอง DDD

จุกทองขอเปรียบเทียบ `__init__` เป็น **"พนักงานต้อนรับและจุดตรวจความปลอดภัย"** ของคลาสครับ
1. **รับของที่จำเป็น (Dependency Injection):** ถ้าร้านกาแฟ (Entity) จำเป็นต้องใช้เครื่องสุ่มเมนู หรือสมุดจดคิว (Repository) เราจะบังคับให้ส่งเข้ามาผ่าน `__init__` เสมอ คลาสจะได้ไม่ต้องไปแอบสร้างระบบพวกนี้ขึ้นมาเองข้างใน
2. **เซ็ตค่าเริ่มต้น (Initial State):** เป็นจุดที่เรารับ **Value Object** เข้ามาเพื่อประกอบร่างให้ Entity พร้อมทำงาน และรับประกันว่า Entity จะเกิดมาในสภาพที่สมบูรณ์ 100% (ไม่มีค่าว่าง หรือค่าที่ผิดกฎธุรกิจแน่นอน)

---

### 💻 ตัวอย่างโค้ด: `__init__` ที่ถูกต้องตามหลัก DDD

มาดูตัวอย่างการสร้าง Entity ใบสั่งกาแฟ (`Order`) กันครับ เราจะไม่รับ `str` หรือ `float` เปล่าๆ เข้ามา แต่เราจะรับ Value Object และ Service ที่จำเป็นเข้ามาทาง `__init__` ครับ

```python
from pydantic import BaseModel, Field
from typing import Callable

# ---------------------------------------------------------
# 1. สร้าง Value Object เพื่อความปลอดภัย
# ---------------------------------------------------------
class OrderId(BaseModel):
    value: str = Field(..., min_length=1)

class OrderStatus(BaseModel):
    value: str = Field(default="PENDING")

# ---------------------------------------------------------
# 2. Entity หลักของเรา
# ---------------------------------------------------------
class Order:
    """
    Entity: ใบสั่งซื้อเครื่องดื่ม
    """
    # ฟังก์ชัน __init__ จะทำงานทันทีเมื่อเราสั่ง order = Order(...)
    def __init__(self, order_id: OrderId, notification_service: Callable[[str], None]):

        # 1. กำหนดสถานะเริ่มต้น (Initial State)
        # ประกอบร่าง Entity ด้วย Value Object เพื่อให้มั่นใจว่าข้อมูลถูกต้องตั้งแต่เกิด
        self.id = order_id
        self.status = OrderStatus(value="PENDING")

        # 2. รับของบริจาค (Dependency Injection)
        # รับบริการแจ้งเตือนจากภายนอกเข้ามาเก็บไว้ในตัว (self) เพื่อรอใช้งาน
        self._notify = notification_service

    def complete_order(self) -> None:
        # Business Logic: เมื่อชงเสร็จ ให้เปลี่ยนสถานะและเรียกใช้ Service ที่ถูกฉีดเข้ามา
        self.status = OrderStatus(value="COMPLETED")

        # เรียกใช้ Service ที่เก็บไว้ใน self
        self._notify(f"Order {self.id.value} is ready!")

# 🕵️‍♂️ บทที่ 1.6: พังเพราะตั้งใจ หรือ พังเพราะทำผิด? (ไขคดี AttributeError)
---

ฮะฮ่า จุกทองเข้าใจความรู้สึกป๋าเลยครับ! ความท้าทายปราบเซียนของการทำ TDD คือการแยกให้ออกนี่แหละครับว่า **"มันแดงเพราะเราเขียนเทสต์ดักไว้ถูกต้องแล้ว (รอให้เราไปเขียน Logic)"** หรือ **"มันแดงเพราะเราเขียนโค้ดในไฟล์เทสต์ผิดเอง"** เคสนี้จัดอยู่ในหมวด **"พังเพราะทำผิดวิธี"** ครับป๋า! ต้นเหตุมาจาก Error ตัวนี้ครับ `AttributeError: 'function' object has no attribute 'is_called_with_choices'` ซึ่งแปลว่า ป๋ากำลังพยายามไปถามหาตัวแปร `is_called_with_choices` จากฟังก์ชันธรรมดาๆ (`def mock_randomly_select():`) ซึ่งในไพธอน ฟังก์ชันปกติมันไม่มีตัวที่เอาไว้จำสถานะ (State) ครับ

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **AttributeError** | แอท-ทริ-บิวต์-เออร์-เรอร์ | **หาคุณสมบัติไม่เจอ** (โปรแกรมพังเพราะเราไปเรียกใช้ตัวแปรที่ไม่มีอยู่จริงใน Object นั้นๆ) |
| **State Verification** | สเตต-เวอ-ริ-ฟิ-เค-ชัน | **การตรวจสอบสถานะ** (การเช็คว่า Mock Object ของเรา ถูกเรียกใช้งานด้วยข้อมูลอะไรบ้าง) |
| **Callable Class / Magic Method** | คอล-เลเบิล คลาส / เมจิก เมท-ธอด | **คลาสที่แปลงร่างเป็นฟังก์ชันได้** (การใช้ `__call__` ทำให้คลาสสามารถถูกเรียกใช้ด้วยวงเล็บ `()` ได้เหมือนฟังก์ชัน) |

### 🛠️ วิธีแก้: อัปเกรดฟังก์ชันจำลอง ให้กลายเป็น "คลาสสายลับ (Mock Class)"

ในคลิปวิดีโอ ช่วงแรกโค้ชเขาอาจจะใช้วิธียัดตัวแปรใส่ฟังก์ชันไปดื้อๆ แต่มันทำให้ IDE แจ้งเตือนและพังได้ง่ายครับ วิธีที่ Clean และเป็น OOP ตามสไตล์สถาปัตยกรรมของเรา คือการสร้าง **Mock Class** เพื่อให้มันจำสถานะ (State) ว่ามันถูกเรียกด้วยข้อมูลอะไรบ้างครับ

ให้ป๋าแก้โค้ดในไฟล์ `tests/test_coffee_shop.py` ตามนี้เลยครับ:

```python

````python
# Coffee_Shop/tests/test_coffee_shop.py
from pydantic import BaseModel, Field

# (สมมติว่าป๋ามี Value Objects เหล่านี้อยู่แล้วตามหลัก DDD)
class DrinkName(BaseModel):
    value: str = Field(..., min_length=1)

class MoneyTHB(BaseModel):
    value: float = Field(..., ge=0.0)

# ==========================================
# 1. สร้างคลาสหน้าม้า (Mock Class) เพื่อเก็บสถานะการถูกเรียกใช้งาน
# ==========================================
class MockRandomlySelect:
    def __init__(self):
        # เตรียมตัวแปร (Attribute) ไว้บันทึกว่า โดนส่งรายชื่อเมนูอะไรมาให้สุ่มบ้าง
        self.is_called_with_choices: list[str] = []

    # Magic Method __call__ ทำให้คลาสนี้ถูกใช้งานแบบฟังก์ชันได้ เช่น mock_obj(['Tea', 'Coffee'])
    def __call__(self, choices: list[str]) -> str:
        # บันทึก State (จดจำไว้ว่าถูกเรียกด้วยอะไร)
        self.is_called_with_choices = choices

        # Business Logic ของเทสต์: บังคับผลลัพธ์ให้ตอบ 'Tea' เสมอ เพื่อให้ Test คาดเดาผลลัพธ์ได้
        return 'Tea'

# ==========================================
# 2. ปรับปรุง Test Case ให้ใช้ Mock Class
# ==========================================
def test_should_randomly_select_a_drink_between_tea_and_coffee_when_buying_surprise():
    # [Arrange] เตรียมระบบ
    # ถอดร่าง Mock Class ออกมาเป็นอ็อบเจกต์หน้าม้า
    mock_selector = MockRandomlySelect()

    # ฉีด (Inject) หน้าม้าเข้าไปใน Entity Shop
    shop = Shop(randomly_select=mock_selector)

    # ห่อหุ้มข้อมูลด้วย Value Object ก่อนเข้าสู่ Domain
    order = DrinkName(value='Surprise')
    pay = MoneyTHB(value=100.0)

    # [Act] กระทำ Business Logic
    # สมมติว่าเมื่อสั่ง 'Surprise' ข้างใน shop.buy() จะมีการเรียกใช้ self.randomly_select(['Tea', 'Coffee'])
    shop.buy(order, pay)

    # [Assert] ตรวจสอบสถานะการทำงาน (State Verification)
    # พิสูจน์ว่า Shop ได้ทำการส่งลิสต์เมนู ['Tea', 'Coffee'] ไปให้ตัวสุ่มทำงานจริงๆ
    assert mock_selector.is_called_with_choices == ['Tea', 'Coffee']
````

💡 หัวใจสำคัญของการหลอกระบบ (Mocking):
การใช้ Class มาทำ Mock จะช่วยให้เราเก็บข้อมูลประวัติการทำงานไว้ใน self.xxx ได้อย่างปลอดภัยและตรวจสอบง่ายครับ ซึ่งต่างจากฟังก์ชันที่พอล็อกิกจบ ตัวแปรข้างในก็หายวับไปหมดเลย นี่แหละครับเคล็ดลับที่ทำให้เราดักจับพฤติกรรมของระบบได้อย่างแม่นยำ!

# 🐛 บทที่ 1.7: จับโป๊ะ Type Mismatch (เมื่อ List ปลอมตัวมา)
---

ฮะฮ่า! มาถึงจุดนี้ป๋าไม่ต้องตกใจครับ อาการแบบนี้เรียกกันในวงการว่า **"ส่งของผิดสเปก"** ครับ! จุกทองเห็นตัวการชัดเจนเลยจาก Error บรรทัดนี้ `drink_name = ['Tea', 'Coffee']` ครับ

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **AttributeError** | แอท-ทริ-บิวต์-เออร์-เรอร์ | **หาคุณสมบัติไม่เจอ** (พังเพราะไปเรียกใช้ `.value` กับของที่ไม่มีสิ่งนี้อยู่) |
| **Type Mismatch** | ไทป์-มิส-แมตช์ | **ชนิดข้อมูลไม่ตรงกัน** (ฟังก์ชันต้องการ Value Object แต่ดันส่ง List เข้าไปให้) |
| **Strong Typing** | สตรอง-ไทป์-ปิง | **การบังคับชนิดข้อมูล** (ข้อดีของการใช้ Value Object คือพอมันจับได้ว่าของปลอมส่งมา มันจะระเบิดตัวเองทันที ไม่ปล่อยให้ทำงานต่อ) |



---

### 🕵️‍♂️ แกะรอยหาคนร้ายใน `shop_logic.py`

ฟังก์ชัน `_validate_drink(self, drink_name: DrinkName)` ของป๋านั้นเขียนดักไว้ดีมากครับ มันตั้งใจรอรับ **Value Object** ที่มีหน้าตาแบบนี้ `DrinkName(value='Tea')` เพื่อที่มันจะได้เจาะไข่แดงเอาค่าข้างในด้วยคำสั่ง `drink_name.value` ได้

**แต่สิ่งที่เกิดขึ้นตอนป๋าสั่ง Surprise คือ:**
ป๋าน่าจะเผลอส่งตัวแปรที่เป็นลิสต์เมนู `['Tea', 'Coffee']` โยนเข้าไปในฟังก์ชัน `_validate_drink()` ตรงๆ เลย (อาจจะเบลอตอนกำลังเรียกใช้ฟังก์ชันสุ่มแน่ๆ ครับ!) ไพธอนมันก็เลยงงว่า *"เอ๊ะ! ลิสต์มันไม่มี `.value` ให้เรียกนี่นา"* แล้วก็เลยสาด Error แดงๆ ใส่หน้าจอครับ

### 🛠️ วิธีแก้: ลำดับความคิดใหม่ตอนสั่ง "Surprise"
