# ☕ บทที่ 1: TDD ร้านกาแฟ (สเต็ป Red Phase สร้างเทสต์ให้พัง)
---

ในสเต็ปนี้เราจะทำตามคลิปเป๊ะๆ โดยแยกไฟล์ระหว่าง **โค้ดทดสอบ (Test)** และ **โค้ดใช้งานจริง (Logic)** ออกจากกัน การทำ TDD เราต้องเริ่มจากเขียนเทสต์ก่อนเสมอครับ

| คำศัพท์ (Technical Term) | คำอ่านไทย (Pronunciation) | ความหมาย (Translation/Definition) |
| :--- | :--- | :--- |
| **Test Case** | เทสต์-เคส | **กรณีทดสอบ** (เงื่อนไขที่เราสร้างขึ้นมาเพื่อพิสูจน์ว่าระบบทำงานถูกต้อง) |

| **Red Phase** | เรด-เฟส | **สถานะพัง** (จุดเริ่มต้นของ TDD คือการรันเทสต์ให้พัง เพื่อให้แน่ใจว่าเทสต์ทำงานได้จริง) |

# 🧱 บทที่ 1.3: สร้างรากฐานด้วย Value Object (ควรเขียนตอนไหน?)
---

คำถามนี้คมมากครับป๋า! ในคลิปวิดีโอที่เขาสอน เขาไม่ได้เขียน Value Object (VO) เพราะโจทย์เน้นความเข้าใจ TDD แบบพื้นฐานที่สุด จึงใช้ Primitive Type (`str`, `float`) ไปก่อน แต่สำหรับโลกของ **DDD (Domain-Driven Design)** และ **Clean Architecture** ที่ป๋ากำลังฝึกปรืออยู่นั้น...

คำตอบคือ: **"เราควรเขียน Value Object เป็นอันดับแรกสุด (Step 0) ก่อนที่จะไปเขียน Entity ครับป๋า!"**

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **Value Object (VO)** | แวลู อ็อบเจกต์ | **อ็อบเจกต์ห่อหุ้มข้อมูล** (ไม่มี ID ของตัวเอง สนใจแค่ค่าที่อยู่ข้างใน และต้องตรวจสอบความถูกต้องของตัวเองได้) |
| **Primitive Obsession** | พริม-มิ-ทีฟ ออบ-เซส-ชัน | **การเสพติดตัวแปรพื้นฐาน** (Anti-pattern ที่ใช้ `int`, `str`, `float` มาแทนคอนเซปต์ทางธุรกิจ ทำให้โค้ดพังง่าย) |
| **Encapsulation** | เอ็น-แคป-ซู-เล-ชัน | **การแคปซูลข้อมูล** (การห่อหุ้มข้อมูลและเงื่อนไขการตรวจสอบไว้ด้วยกันในที่เดียว) |

---

### 🤔 ทำไมต้องสร้าง Value Object ก่อน Entity?

ถ้าป๋าไม่สร้าง VO ก่อน ภาระการตรวจสอบข้อมูลทั้งหมด (เช่น เช็คว่าเงินติดลบไหม, เช็คว่าชื่อเครื่องดื่มว่างเปล่าไหม) จะไปตกหนักอยู่ที่ Entity (`Cafe`) ทำให้ Entity บวมและโค้ดรกครับ

ในทางกลับกัน ถ้าเราสร้าง VO ขึ้นมาทำหน้าที่ "เฝ้าประตู" ไว้ตั้งแต่แรก Entity ของเราจะสนใจแค่ **Business Logic หลัก** (เช่น การหักลบเงินทอน) โดยไม่ต้องกังวลว่าข้อมูลที่รับเข้ามาจะผิดพลาดเลยครับ

### 🛠️ TDD Step-by-Step: การสร้าง Value Object `Money` ด้วย Pydantic

เราจะมาประยุกต์ใช้หลักการ TDD (Red-Green-Refactor) เพื่อสร้าง `Money` กันก่อนเลยครับ

```python

# ไฟล์: tests/test_value_objects.py
import pytest
from pydantic import ValidationError
# from domain.value_objects import Money  <-- เดี๋ยวเราจะไปสร้างตัวนี้กัน

# ---------------------------------------------------------
# 1. เขียน Test ให้ Value Object (Red Phase)
# ---------------------------------------------------------

def test_money_cannot_be_negative():
    # [Arrange & Act & Assert]
    # คาดหวังว่าถ้าเผลอสร้างเงิน "ติดลบ" ระบบจะต้องเตะ Error (ValidationError) ออกมาทันที
    with pytest.raises(ValidationError):
        invalid_money = Money(amount=-10.0)

def test_money_can_be_created_with_valid_amount():
    # [Arrange & Act] สร้างเงิน 50 บาท
    valid_money = Money(amount=50.0)

    # [Assert] ค่า amount ต้องเท่ากับ 50.0 เป๊ะๆ
    assert valid_money.amount == 50.0

````python
# ไฟล์: domain/value_objects.py
from pydantic import BaseModel, Field

# ---------------------------------------------------------
# 2. สร้าง Value Object ให้ Test ผ่าน (Green Phase)
# ---------------------------------------------------------

class Money(BaseModel):
    """
    Value Object สำหรับจัดการเรื่องเงิน
    Business Logic: จำนวนเงินในระบบห้ามมีค่าติดลบเด็ดขาด (ge=0.0 คือ Greater than or Equal to 0)
    """
    amount: float = Field(..., ge=0.0, description="จำนวนเงินต้องไม่ติดลบ")

class DrinkName(BaseModel):
    """
    Value Object สำหรับชื่อเครื่องดื่ม
    Business Logic: ห้ามเป็น String ว่างๆ ต้องมีตัวอักษรอย่างน้อย 1 ตัว
    """
    value: str = Field(..., min_length=1, description="ชื่อเครื่องดื่มห้ามว่าง")
````

````python
# ไฟล์: domain/logic.py
# from domain.value_objects import DrinkName, Money

class Cafe:
    def __init__(self):
        # ราคาสินค้าก็ควรเก็บเป็น Value Object เช่นกัน
        self.PRICE = {
            "Tea": Money(amount=20.0),
            "Coffee": Money(amount=30.0)
        }

    # ห้ามใช้ drink: str, money: float ตรงๆ! ต้องบังคับรับเป็น Value Object
    def buy(self, drink: DrinkName, payment: Money) -> tuple[DrinkName, Money]:

        drink_price = self.PRICE.get(drink.value)
        if not drink_price:
            raise ValueError("Drink not in menu")

        # 1. Guard Clause: เช็คเงินพอไหม? (ข้อมูลที่เข้ามาการันตีแล้วว่าไม่ติดลบ)
        if payment.amount < drink_price.amount:
            raise ValueError("Money not enough")

        # 2. Calculate: คำนวณเงินทอน
        change_amount = payment.amount - drink_price.amount

        # 3. ห่อกลับเป็น Value Object ก่อนส่งคืนออกไป
        change = Money(amount=change_amount)

        return drink, change
````

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **Syntax Error** | ซิน-แท็กซ์ เออร์-เรอร์ | **ไวยากรณ์ผิดพลาด** (การพิมพ์โค้ดผิดกฎของภาษา เช่น ลืมพิมพ์ `def`, ลืมเครื่องหมาย `:`) |
| **Typo** | ไท-โป | **พิมพ์ผิด/พิมพ์ตกหล่น** (เรื่องธรรมชาติที่โปรแกรมเมอร์ทุกคนต้องเจอทุกวัน) |
| **Absolute Import** | แอ๊บ-โซ-ลูท อิม-พอร์ต | **การนำเข้าไฟล์แบบระบุตำแหน่งเต็ม** (เช่น `from domain.value_object import ...`) |

### 🚀 สเต็ปแก้ปัญหา: ก๊อปปี้ทับไปเลยครับป๋า!

เพื่อให้ป๋าไปต่อได้แบบลื่นๆ จุกทองเตรียมโค้ดที่ถูกต้องสมบูรณ์แบบของทั้ง 3 ไฟล์มาให้แล้วครับ ป๋าแค่ก๊อปปี้ไปวางทับในไฟล์ของป๋าได้เลยครับ โดยยึดโครงสร้างโฟลเดอร์ตามนี้ครับ:

`<โครงสร้างโฟลเดอร์>`
```text
Coffee_Shop/
├── domain/
│   ├── __init__.py
│   ├── shop_logic.py     # ไฟล์ที่ 2
│   └── value_object.py   # ไฟล์ที่ 1
└── tests/
    ├── __init__.py

📁 ไฟล์ที่ 1: domain/value_object.py
(ไฟล์นี้เก็บ Value Object ที่คอยปกป้องข้อมูลให้เรา)

````python
from pydantic import BaseModel, Field

class DrinkName(BaseModel):
    """Value Object สำหรับชื่อเครื่องดื่ม"""
    value: str = Field(..., min_length=1)

class MoneyTHB(BaseModel):
    """Value Object สำหรับเงินบาท (ห้ามติดลบ)"""
    amount: float = Field(..., ge=0.0)
````


📁 ไฟล์ที่ 2: domain/shop_logic.py
(ก๊อปปี้ไปทับไฟล์เดิมที่พังอยู่ได้เลยครับ แก้คำผิดและ Import ให้ครบแล้ว)

````python

# นำเข้า Value Object ที่สร้างไว้มาใช้งาน
from domain.value_object import DrinkName, MoneyTHB

class Shop:
    def __init__(self):
        # กำหนดราคาสินค้าโดยห่อด้วย Value Object
        self.PRICE = {
            'Tea': MoneyTHB(amount=20.0),
            'Coffee': MoneyTHB(amount=30.0)
        }

    def buy(self, drink: DrinkName, payment: MoneyTHB) -> tuple[DrinkName, MoneyTHB]:
        # ดึงราคาจาก Dictionary
        drink_price = self.PRICE.get(drink.value)

        # Guard Clause 1: เช็คว่ามีเมนูนี้ไหม
        if drink_price is None:
            raise ValueError("Drink not in menu")

        # Guard Clause 2: เช็คว่าเงินพอไหม
        if payment.amount < drink_price.amount:
            raise ValueError("Money not enough")

        # คำนวณเงินทอน (Business Logic หลัก)
        change_amount = payment.amount - drink_price.amount

        # ห่อผลลัพธ์กลับเป็น Value Object ก่อนคืนค่า
        return drink, MoneyTHB(amount=change_amount)
````


📁 ไฟล์ที่ 3: tests/test_coffee_shop.py
(ไฟล์นี้สำหรับเขียน Test Case เพื่อรันพิสูจน์ระบบครับ)

````python
from domain.shop_logic import Shop
from domain.value_object import DrinkName, MoneyTHB

def test_should_return_tea_and_change_30_when_buy_tea_with_50():
    # [Arrange] เตรียมระบบ
    shop = Shop()
    order = DrinkName(value="Tea")
    pay = MoneyTHB(amount=50.0)

    # [Act] กระทำ Business Logic: สั่งซื้อ
    drink, change = shop.buy(order, pay)

    # [Assert] ตรวจสอบผลลัพธ์
    assert drink.value == "Tea"
    assert change.amount == 30.0
````

# 🎯 บทที่ 1.4: สร้าง Custom Exception แบบต้นฉบับเป๊ะๆ
---

ตาไวมากครับป๋า! จุกทองเห็นจากรูปที่ 3 แล้ว ต้นฉบับเขาใช้เทคนิคการสร้าง **Custom Exception** (`DrinkNotInMenu`) และแยกการตรวจสอบไปไว้ในฟังก์ชันส่วนตัว (Private Method) ที่ชื่อ `_validate_drink` เพื่อให้โค้ดในฟังก์ชัน `buy` อ่านง่ายและเป็นระเบียบที่สุดครับ

การทำแบบนี้ดีมากๆ เพราะเวลาที่ระบบ Error มันจะไม่ฟ้องแค่ `ValueError` ทั่วไป แต่จะฟ้องชื่อ Error ที่สื่อความหมายทางธุรกิจชัดเจนเลยครับ เรามาปรับโค้ดของเราให้เหมือนต้นฉบับ โดยยังคงความแข็งแกร่งของ Value Object ไว้กันครับ!

| Technical Terms | คำอ่านภาษาไทย | ความหมาย / คำแปล |
| :--- | :--- | :--- |
| **Custom Exception** | คัส-ตอม เอ็ก-เซป-ชัน | **ข้อผิดพลาดที่สร้างขึ้นเอง** (การสร้าง Class Error เฉพาะเจาะจงสำหรับ Business ของเรา) |
| **Private Method** | ไพร-เวต เมท-ธอด | **ฟังก์ชันใช้งานภายใน** (ฟังก์ชันที่ขึ้นต้นด้วย `_` เพื่อบอกว่าฟังก์ชันนี้ใช้เรียกกันเองภายใน Class เท่านั้น ห้ามคนนอกเรียกใช้) |
| **pytest.raises** | พาย-เทสต์ เรส-เซส | **การดักจับ Error ใน Test** (คำสั่งสำหรับทดสอบว่าโค้ดของเราทำการพ่น Error ออกมาตามที่คาดหวังหรือไม่) |

> **💡 ข้อคิดจากจุกทอง:**
> การตั้งชื่อ Exception ให้สื่อถึง Business Logic (เช่น `DrinkNotInMenu` หรือ `MoneyNotEnough`) คือหัวใจสำคัญของ DDD ครับ มันทำให้โค้ดของเรากลายเป็นเอกสาร (Self-documenting code) อ่านปุ๊บรู้ปั๊บว่าพังเพราะกฎข้อไหน!

### 🛠️ สเต็ป 1: สร้าง Custom Exception และปรับ Logic ให้เหมือนต้นฉบับ

ป๋าก๊อปปี้โค้ดด้านล่างนี้ ไปทับในไฟล์ `domain/shop_logic.py` ได้เลยครับ จุกทองเพิ่ม Class `DrinkNotInMenu` และฟังก์ชัน `_validate_drink` ให้เหมือนในคลิปเป๊ะๆ แล้วครับ


````python

# ไฟล์: domain/shop_logic.py
from domain.value_object import DrinkName, MoneyTHB

# ---------------------------------------------------------
# สร้าง Custom Exception (สืบทอดจาก Exception พื้นฐานของ Python)
# ---------------------------------------------------------
class DrinkNotInMenu(Exception):
    """Error: เมื่อลูกค้าสั่งเครื่องดื่มที่ไม่มีในเมนู"""
    pass

class MoneyNotEnough(Exception):
    """Error: เมื่อลูกค้าจ่ายเงินไม่พอ"""
    pass

# ---------------------------------------------------------
# Entity: Shop
# ---------------------------------------------------------
class Shop:
    def __init__(self):
        self.PRICE = {
            'Tea': MoneyTHB(amount=20.0),
            'Coffee': MoneyTHB(amount=30.0)
        }

    def _validate_drink(self, drink: DrinkName):
        # ตรวจสอบว่าชื่อเครื่องดื่ม (drink.value) มีอยู่ใน Dictionary หรือไม่
        if drink.value not in self.PRICE:
            # พ่น Custom Exception ตามต้นฉบับ
            raise DrinkNotInMenu()

    def buy(self, drink: DrinkName, payment: MoneyTHB) -> tuple[DrinkName, MoneyTHB]:
        # 1. เรียกใช้งาน Private Method เพื่อเช็คเมนู (ถ้าไม่มี จะเด้ง Error ออกไปเลย)
        self._validate_drink(drink)

        # 2. ดึงราคาออกมาคำนวณ (ถึงบรรทัดนี้มั่นใจได้ว่ามีเมนูแน่นอน)
        drink_price = self.PRICE[drink.value]

        # 3. Guard Clause: เช็คเงินพอไหม?
        if payment.amount < drink_price.amount:
            raise MoneyNotEnough()

        # 4. คำนวณเงินทอน และห่อกลับเป็น Value Object
        change_amount = payment.amount - drink_price.amount
        return drink, MoneyTHB(amount=change_amount)
````
